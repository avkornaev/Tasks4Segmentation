# Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

# Settings

```python
imgNo = 0
NoOfClasses = 2
fig_size = (10,10)
IMG_DIR = "/WRITE_ACCESS_STORAGE/ToyDataset4Segmentation/train_images/"
ANN_DIR = "/WRITE_ACCESS_STORAGE/ToyDataset4Segmentation/train_masks/"
```

# Download an image

```python
img_list = os.listdir(IMG_DIR)
ann_list = os.listdir(ANN_DIR)

img_path = IMG_DIR+img_list[0]
ann_path = ANN_DIR+img_list[0]

print(img_list, ann_list, sep='\n')
```

```python
img = Image.open(img_path)
ann = Image.open(ann_path)
#img = cv2.cvtColor(cv2.imread(str(files[0]), cv2.IMREAD_GRAYSCALE), cv2.COLOR_BAYER_RG2RGB)
img
```

```python
img = np.array(img)
ann = np.array(ann)
```

```python
img.dtype, ann.dtype
```

```python
print(ann.min(), ann.max(), ann.shape, img.shape)
```

```python
classN = [0]*NoOfClasses

for i in range(NoOfClasses):
    sub = ann.copy()
    sub[sub!=(i+1)] = 0
    sub[sub!=0] = 1
    classN[i]=sub
```

```python
plt.figure(figsize = (fig_size[0],fig_size[1]))
for i in range(NoOfClasses):
    maskedImage = img*classN[i]
    plt.subplot(1,NoOfClasses,i+1)
    plt.imshow(maskedImage)
    plt.title('Image with mask of class'+ str(i))
    print(maskedImage.shape)
    
```
